### Feature and target

```python
X = df.drop(columns= ["hsc_result", "date"])
y = df["hsc_result"]
```

```python
X
```

```python
y
```

```python
numerical_feature = X.select_dtypes(include=["int64", "float64"]).columns 

```

```python
numerical_feature
```

```python
categorical_features = X.select_dtypes(include=["object"]).columns
```

```python
categorical_features
```

```python
num_transformer = Pipeline(steps=[("imputer", SimpleImputer(strategy='median')), ("scaler", StandardScaler())])

```

```python
num_transformer
```

```python
cat_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")), 
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
) 
```

```python
cat_transformer
```

```python
preprocessor = ColumnTransformer(
    transformers = [
        ("num", num_transformer, numerical_feature),
        ("cat", cat_transformer, categorical_features)
    ]
)
```

```python
preprocessor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
) 
```

```python
X_train
```

```python
X_test
```

### linear regression using pipeline

```python
lr_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", LinearRegression())
    ]
)
lr_pipeline.fit(X_train,y_train)
```

### SGD

```python
ETA = 0.001 
MAX_ITER = 3000 
ALPHA = 0.0001 

sgd_pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor), 
        ("model", SGDRegressor(
            loss= "squared_error", 
            penalty= 'l2',
            alpha = ALPHA, 
            learning_rate= 'constant', 
            eta0= ETA, 
            max_iter= MAX_ITER, 
            random_state= 42
        ))
    ]
)
sgd_pipeline.fit(X_train, y_train)
```

```python
# Evaluation  function
def evaluate(model, name):
    train_pred = model.predict(X_train) 
    test_pred = model.predict(X_test) 

    print(f"Name of the model {name}") 
    print("---"*40) 

    print("Train R2 :", round(r2_score(y_train, train_pred), 4))
    print("Test R2 :", round(r2_score(y_test, test_pred), 4)) 
    print("RMSE :", round(np.sqrt(mean_squared_error(y_test, test_pred)), 4))
    print("MAE :", round(mean_absolute_error(y_test, test_pred), 4))

    return (
        r2_score(y_test, test_pred),
        np.sqrt(mean_squared_error(y_test, test_pred)),
        mean_absolute_error(y_test, test_pred)
    )

```

```python
lr_r2, lr_rmse,lr_mase = evaluate(
    lr_pipeline, 
    "Linear Regression"
)
```

```python
sgd_r2, sgd_rmse,sgd_mase = evaluate(
    sgd_pipeline, 
    "SGD Regression"
)
```

### Cross validation

```python
cv_scores = cross_val_score(
    lr_pipeline, 
    X, 
    y, 
    cv = 5, 
    scoring= 'r2' 
)

print("Scores", cv_scores) 
print("Average Cv 42", cv_scores.mean())
```
