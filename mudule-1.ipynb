## 1.5 Univariate Analysis (Categorical Value Analysis -1)

```python
sns.countplot(data=df, x=df['Survived'])
```

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts() / len(df) * 100
```

```python
survived_count = df['Survived'].value_counts()

plt.pie(survived_count, autopct="%1.1f")
```

```python
survived_count = df['Survived'].value_counts()
survived_label = df['Survived'].unique()

print(survived_label)
plt.pie(survived_count,labels = survived_label, autopct="%1.1f")
```

```python
survived_count = df['Survived'].value_counts()
survived_label = df['Survived'].unique()

print(survived_label)

plt.title('Pie chart of survival rate')
plt.pie(survived_count,labels = survived_label, autopct="%1.1f")
plt.show()
```

## 1.7 Univariate Analysis (Numerical Column Analysis - 1)

```python
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

sns.histplot(df['Age']) 
plt.show()
```

```python
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

sns.histplot(df['Age'],bins=50) 
plt.show()
```

```python
plt.xlabel('Ages')
plt.ylabel('Probabilty Distribution')
plt.title('KDE of Age')

sns.kdeplot(df['Age'])
plt.show()
```
